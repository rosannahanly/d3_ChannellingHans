```python
import pandas as pd
import numpy as np
```

```python
gci = pd.read_csv("GCI_CompleteData4.csv", sep=",")
gci
```

```python
# Fills null values for global competitiveness index with the mean of the other values of GCI for that country

gci.loc[gci.Global_Competitiveness_Index.isnull(), 'Global_Competitiveness_Index'] = gci.groupby('Country').Global_Competitiveness_Index.transform('mean')
```

```python
gci
```

```python
gci.to_csv("GCI4_imputed_values.csv", sep=',', index=False)
```
